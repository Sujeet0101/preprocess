```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_excel("hostel_excel.xlsx")
dataset.head()
```

```python
dataset = dataset[['Hid','HostelName','District','Place/City','HostelType','Facilities']]
```

```python
dataset.head()
```

```python
dataset['District'] = dataset['District'].apply(lambda x:x.lower())
```

```python
dataset['Place/City'] = dataset['Place/City'].apply(lambda x:x.lower())
```

```python
dataset['HostelType'] = dataset['HostelType'].apply(lambda x:x.lower())
```

```python
dataset['Facilities'] = dataset['Facilities'].apply(lambda x:x.lower())
```

```python
dataset.head()
```

```python
dataset['HostelName'] = dataset['HostelName'].apply(lambda x:x.lower())
```

```python
dataset.head()
```

```python
#Create the list of all featues for each hostel
features = ["HostelName","District","Place/City","HostelType","Facilities"]
all_features = []
for index, row in dataset.iterrows():
    hostel_features = []
    for feature in features:
          hostel_features.append(row[feature])
    all_features.append(" ".join(hostel_features))

```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Use CountVectorizer to convert the list of features into a matrix of token counts
vectorizer = CountVectorizer()
feature_matrix = vectorizer.fit_transform(all_features)
```

```python
# Define a function to get recommendations based on user input
def get_recommendations(name, district, placeorcity, hosteltype, facilities):
  # Create a list of the user's input
  input_features = [name, district, placeorcity, hosteltype, facilities]
  # Use CountVectorizer to convert the input into a matrix of token counts
  input_matrix = vectorizer.transform([" ".join(input_features)])
  # Calculate the cosine similarity between the input and all hostels
  similarity = cosine_similarity(input_matrix, feature_matrix)
  # Get the indices of the most similar hostels
  most_similar = similarity.argsort()[:, ::-1]
  # Return a list of the most similar hostels
  return dataset.iloc[most_similar]
```

```python
similar_hostels = get_recommendations(name="friends boys hostel", district="bhaktapur", placeorcity="dokchee", hosteltype="boys", facilities="wifi")
```
